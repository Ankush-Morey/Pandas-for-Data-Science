```python
#Missing Values
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df5 = pd.DataFrame(people)
```

```python
df5
```

```python
#Drop NA rows
df5.dropna()
```

```python
#There are two parameters in this method, how = 'any', if any attribute is NA it will remove that row. This is default 
df5.dropna(axis='index',how='any')
```

```python
#There are two parameters in this method, how = 'all', if all attributes in subset are NA it will remove that row.
df5.dropna(axis='index',how='all', subset=['last','email'])
```

```python
#to check if NA or not
df5.isna()
```

```python
#Replace NA with any value
df5.fillna(0)
```

```python
#Casting
#Check datatype..dtypes is not a method, strings are represented as objects
df5.dtypes
```

```python
#type of nan is float, to get mean of age we cast age to float
df5['age'].replace('Missing',np.nan,inplace=True)
df5['age']=df5['age'].astype(float)
```
